```python
import pandas as pd
from textblob import TextBlob

import textstat
import spacy
import nltk
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv("/home/ruchika/Downloads/test.csv", encoding="utf-8")
```

```python
df = df.assign(final=lambda x: x['Title'].astype(str) + x['Text'].astype(str))
df= df.drop(['Text','Title'],axis=1)
```

```python
df = df.rename(columns={'final': 'Text'})
df
```

```python
def tokenize_text(text):
    return nltk.word_tokenize(text)

# Tokenize the text in the DataFrame
df['tokenized_text'] = df['Text'].apply(tokenize_text)




```

```python
tokens = df['Text'].apply(nltk.word_tokenize)
```

```python

# Perform POS tagging
tagged = tokens.apply(nltk.pos_tag)

# Find personal pronouns
pronouns = tagged.apply(lambda x: [word for word, tag in x if tag == 'PRP'])

# Add the personal pronouns to the dataframe
df['Personal Pronouns'] = pronouns

# Print the updated dataframe
print(df.head())
```

```python
# Calculate the positive and negative scores using TextBlob
df['Positive Score'] = df['Text'].apply(lambda x: TextBlob(x).sentiment.polarity if x.strip() else 0)
df['Negative Score'] = df['Text'].apply(lambda x: 1 - TextBlob(x).sentiment.polarity if x.strip() else 0)

```

```python
# Data Preprocessing
df['Text'] = df['Text'].apply(lambda x: " ".join(x.lower() for x in x.split())) # Convert to lowercase
df['Text'] = df['Text'].str.replace('[^\w\s]','') # Remove punctuation
stop_words = set(nltk.corpus.stopwords.words('english'))
df['Text'] = df['Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop_words)) # Remove stopwords

# Sentiment Analysis
df['polarity'] = df['Text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['Text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Print the updated dataframe
print(df.head())
```

#  Calculate the average sentence length using textstat

```python
df['Avg Sentence Length'] = df['Text'].apply(lambda x: textstat.avg_sentence_length(x) if x.strip() else 0)
```

```python
df
```

```python
# Calculate the percentage of complex words using textstat
df['Percentage of Complex Words'] = df['Text'].apply(lambda x: textstat.difficult_words(x)/textstat.lexicon_count(x)*100 if x.strip() else 0)
```

```python
# Calculate the FOG index using textstat
df['FOG Index'] = df['Text'].apply(lambda x: textstat.gunning_fog(x) if x.strip() else 0)
```

```python
# Calculate the average numbefrom textblob.blob import TextBlobr of words per sentence using textstat
df['Avg Number of Words per Sentence'] = df['Text'].apply(lambda x: textstat.avg_sentence_length(x) if x.strip() else 0)
```

```python
# Calculate the complex word count using textstat
df['Complex Word Count'] = df['Text'].apply(lambda x: textstat.difficult_words(x) if x.strip() else 0)

# Calculate the word count using textstat
df['Word Count'] = df['Text'].apply(lambda x: textstat.lexicon_count(x) if x.strip() else 0)

# Calculate the average syllables per word using textstat
df['Syllable per Word'] = df['Text'].apply(lambda x: textstat.syllable_count(x)/textstat.lexicon_count(x) if x.strip() else 0)
```

```python
df
```

```python
# Tokenize the text and calculate the length of each word
word_lengths = df['Text'].apply(lambda x: [len(word) for word in x.split()])

# Calculate the average word length for each row
avg_word_lengths = word_lengths.apply(lambda x: sum(x) / len(x))

# Add the average word length to the dataframe
df['Average Word Length'] = avg_word_lengths

# Print the updated dataframe
print(df.head())
```
